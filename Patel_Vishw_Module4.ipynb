# Step 2: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, average_precision_score, classification_report, confusion_matrix
from autogluon.tabular import TabularPredictor
```

# Step 3: Load the Dataset

```python
# Load the dataset
df = pd.read_csv("/Users/vishwpatel/Desktop/EAI 6020/Module 4/creditcard.csv")

# Print class distribution (to show imbalance)
print("Class distribution:")
print(df['Class'].value_counts(normalize=True))
```

# Step 4: Train AutoML Model with AutoGluon

```python
# Define target label
label = 'Class'

# Fit the AutoML model using TabularPredictor
predictor = TabularPredictor(label=label, eval_metric='recall', verbosity=2).fit(df)
```

# Step 5: Predict Probabilities

```python
# Get predicted fraud probabilities (Score between 0 and 1)
y_true = df[label]
y_scores = predictor.predict_proba(df)[1]  # Column 1 = probability of fraud (Class = 1)
```

# Step 6: Plot Precision-Recall Curve

```python
# Calculate precision-recall curve
precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
avg_precision = average_precision_score(y_true, y_scores)

# Plot the curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label='PR Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title(f'Precision-Recall Curve (Avg Precision = {avg_precision:.4f})')
plt.legend()
plt.grid(True)
plt.show()
```

# Step 7: Tune the Classification Threshold

```python
# Set a custom threshold (e.g., 0.3 to increase sensitivity)
threshold = 0.3
custom_predictions = np.where(y_scores >= threshold, 1, 0)
```

# Step 8: Evaluate Performance with the New Threshold

```python
# Print confusion matrix and classification report
print(f"\nEvaluation at threshold = {threshold}")
print("Confusion Matrix:")
print(confusion_matrix(y_true, custom_predictions))

print("\nClassification Report:")
print(classification_report(y_true, custom_predictions, digits=4))
```
